```python
def newsInMonth(data):
    # split the text into title lists
    title_list = re.split("前 ", data)
    title_splited_list = []
    for title in title_list:
        title_first_split = title.split(" ⋅ ")
        publish_date = title_first_split[-1]
        if ('天' in publish_date or "小时" in publish_date or "分钟" in publish_date):
            title_splited_list.append(title.split(" ⋅ ")[0].split(' ')[-2])
    return title_splited_list
usa_title_lst = newsInMonth(usaNews)
uk_title_lst = newsInMonth(ukNews)
```

```python
# GET lexical analysis through the API provided by Baidu
def getWords(title_list):
    words_lst = []
    for title in title_list:
        found = False
        while found == False: # THIS loop sovles the FSP BUG
            response = requests.post(url_lexer,
                                   data=json.dumps({"text":title}),
                                   headers=headers)
            if "error_msg" not in response.json():
                found = True
                words_lst.extend(response.json()["items"])
    return words_lst
usa_words_lst = getWords(usa_title_lst)
uk_words_lst = getWords(uk_title_lst)
```

```python
# TRNASFER words list to dataframe
def getDataFrame(pos, words_lst):
    df = pd.DataFrame(words_lst)[["item", "pos", "ne"]]
    # only keep the nouns
    df = df[df["pos"] == pos]
    return df
usa_DataFrame = getDataFrame("n", usa_words_lst)
uk_DataFrame = getDataFrame("n", uk_words_lst)
```

```python
# COUNT the words and keep the top 10 
count_usa_words = usa_DataFrame["item"].value_counts().head(10)
count_uk_words = uk_DataFrame["item"].value_counts().head(10)
print(count_uk_words.head())
print(count_usa_words.head())
```

```python
# COMBINE the two top 10 used words_list
words = pd.concat([count_usa_words, count_uk_words])
words = words.groupby(by=words.index).sum() # ADD repeated words together
words
```

```python
# GENERATE nodes data
# FIRST two nodes are the basic area data
# category is used to separate heading nodes and normal nodes, 0 and 1 are keys in categories
nodes = [{"name" : "USA", "value" : len(usa_title_lst), 
          "symbolSize" : len(usa_title_lst) * 3, "category" : 0},
         {"name" : "UK", "value" : len(uk_title_lst), 
          "symbolSize" : len(uk_title_lst) * 3, "category" : 0}]
# ADD the words
for index in range(len(words)):
    nodes.append({'name':words.index[index],"value":words.values[index],
                  'symbolSize':words.values[index]*3,'category':1})
nodes
```

```python
# GET relationship data, which displays which country the words is related to
# BOTH "source" and "target" are parameters 
# USA
links_usa = []
for i in range(len(count_usa_words)):
    links_usa.append({"source":'USA',"target":count_usa_words.index[i]})

# UK
links_uk = []
for i in range(len(count_uk_words)):
    links_uk.append({"source":"UK","target":count_uk_words.index[i]})
links = links_uk + links_usa
links
```

```python
# GENERATE catergories
categories=[{"name" : "country"},{"name" : "key words"}]
```

```python
# DRAW the graph

# line_curve : the extent of how curve the line is; graph_layout : "force" or "circular"
graph = pe.Graph("Coronavirus news analysis", width=700,height=700)
graph.add("", nodes, links, categories, is_label_show=True,
         is_roam=False,
         line_curve=0.1,
         graph_layout="circular")
graph
```
