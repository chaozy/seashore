```python
# split the text into title lists
title_list = data.split("前") # each title ends with letter 前
title_list.remove('')
# Deriving elements from titles
title_splited = []
for title in title_list:
    title_first_split = title.split(" ⋅ ")
    title_second_split = title_first_split[0].split(' ')

    dic = {"Title" : title_second_split[-2], "Publisher" : title_second_split[-1], 
           "Comment" : title_first_split[-2], "Publish Date" : title_first_split[-1]}
    title_splited.append(dic)
# Display in table
title_table = pd.DataFrame(title_splited)
```

```python
# CALCULATE publish time
def time_formatt(date):
    if "小时" in date:
        return 0
    elif "天" in date:
        return int(date.split("天")[0])
# CLEAN the data
title_table["Publish Time"] = title_table["Publish Date"].apply(lambda x : time_formatt(x))
# CHANGE Publish Date into Date Format
title_table["Date"] = pd.to_datetime(datetime.today().strftime('%Y-%m-%d')) - \
            title_table["Publish Time"].apply(lambda x : pd.Timedelta(x, unit='d'))
title_table.drop(['Publish Date','Publish Time'],axis = 1,inplace = True)
title_table
```

```python
# CLEAN comment element
title_table["Comment"] = title_table["Comment"].str.replace("评论", '').astype(int)
# astype(datatype) is used for vector while int() is used for scalar values
```

```python
def getSentiment(title):
    try:
        found = False
        while found == False:
            response = requests.post(url_sentiment_classify,
                                   data=json.dumps({"text":title}),
                                   headers=headers)
            if "error_msg" not in response.json():
                found = True
        return response.json()["items"][0]["positive_prob"]
    except:
        return 0
title_table["Positive_prob"] = title_table["Title"].apply(lambda x : getSentiment(x))
title_table
```

```python
# Split titles into groups according to their date
title_groups = title_table.groupby(by = "Date").agg(["mean", "sum"]) # agg method will only consider numerical values
title_groups
```

```python
# DEFINE x and y axises
x = title_groups.index.astype("str")
y_positive = title_groups["Positive_prob"]["mean"]
y_comment = title_groups["Comment"]["sum"]
```

```python
# DRAW the line graph using pyecharts
line = pe.Line("SunYung Public Opinion")
line.add("Positive Prob of Sun Yung", x, y_positive, is_smooth=True, 
        tooltip_axispointer_tyoe="cross", mark_line=["average"], area_opacity=0.3)
bar = pe.Bar("SunYung Public Opinion")
bar.add("Comment of Sun Yung news", x, y_comment, mark_line=["average"])

overlap = pe.Overlap()
overlap.add(bar)
overlap.add(line, yaxis_index=1, is_add_yaxis=True)
```
